```python
import matplotlib.pyplot as plt
import matplotlib
# 한글 설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic'		# window용
# matplotlib.rcParams['font.family'] = 'AppleGothic'			# mac용
# fontsize
matplotlib.rcParams['font.size'] = 13
# 그래프에 - 보이기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
# x축,y축 데이터 - list,tuple 타입, DataFrame데이터
x = [1,2,3]
y = [2,4,8]		# SM 매출액
y2 = [5,1,3]	# LG 매출액
y3 = [1,2,5]	#	SK 매출액

plt.figure(figsize = (10,5))

plt.plot(x,y, label= 'SM매출액', marker='o', ls='--')
plt.plot(x,y2, label = 'LG 매출액', marker='s', ls='-.')
plt.plot(x,y3, label = 'SK 매출액', marker='d')

# 그래프에 값 넣기
# for idx,txt in enumerate(y):
# 	plt.text(x[idx],y[idx]+0.2,txt,ha='center')

for i in range(3):
	plt.text(x[i],y[i]+0.2,y[i],ha='center')
	plt.text(x[i],y2[i]+0.2,y2[i],ha='center')
	plt.text(x[i],y3[i]+0.2,y3[i],ha='center')

plt.title('기업별 매출액')
plt.legend(ncol = 3)		# 범례출력 : 기본-세로출력, ncol = 3 - 가로출력
plt.show()
```
