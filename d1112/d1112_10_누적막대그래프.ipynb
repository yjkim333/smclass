```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
# 한글 설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic'		# window용
# matplotlib.rcParams['font.family'] = 'AppleGothic'			# mac용
# fontsize
matplotlib.rcParams['font.size'] = 13
# 그래프에 - 보이기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_excel('score.xlsx')
df
```

누적 막대 그래프
bottom = 첫 번째 값

```python
# 

plt.figure(figsize=(7,4))

plt.bar(df['이름'],df['국어'],label = '국어')

# bottom = df['국어'] -> df['국어'] 그래프 위로 그림 
plt.bar(df['이름'],df['영어'], bottom=df['국어'], label = '영어')
plt.bar(df['이름'],df['수학'], bottom=df['국어']+df['영어'], label = '영어')

plt.xticks(rotation=45)

plt.legend(ncol=3, loc='upper center')
plt.ylim(0,350)

# 값을 그래프에 입력
for i in range(8):
	plt.text(df['이름'][i], df['국어'][i]-10, df['국어'][i], ha='center')
	plt.text(df['이름'][i], df['국어'][i]+df['영어'][i]-10, df['영어'][i], ha='center')
	plt.text(df['이름'][i], df['국어'][i]+df['영어'][i]+df['수학'][i]-5, df['수학'][i], ha='center')



plt.show()
```
