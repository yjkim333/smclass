```python
import matplotlib.pyplot as plt
import matplotlib
# 한글 설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic'		# window용
# matplotlib.rcParams['font.family'] = 'AppleGothic'			# mac용
# fontsize
matplotlib.rcParams['font.size'] = 13
# 그래프에 - 보이기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
import pandas as pd
df = pd.read_excel('score.xlsx',index_col='지원번호')
df
```

```python
# x축 - 지원번호, y축 - 국어점수
# 그래프 선 색 - red, 마커 색 - yellow, 마커테두리 - blue, 선 타입 - --

x = df.index
k = df['국어']
e = df['영어']
m = df['수학']

# 그래프 크기조절
# figsize=(x,y)
# dpi - 그래프해상도 조절 - 해상도가 높을 수록 이미지 크기 커짐
# facecolor - 배경색 설정
plt.figure(figsize=(10,6),facecolor='gray')		# x축 5, y축 5 사이즈 조절

plt.plot(x,k, color = 'red', marker = 'o', ms = 6, mec = 'blue', mfc = 'yellow', ls = '-', label = '국어점수')
plt.plot(x,e, color = 'green', marker = 'o', ms = 6, mec = 'blue', mfc = 'black', ls = ':', label = '영어점수')
plt.plot(x,m, color = 'blue', marker = 'o', ms = 6, mec = 'blue', mfc = 'red', ls = '-.', label = '수학점수')

plt.title('성적그래프',size=10)
plt.xlabel('학생번호', color = 'gray', size=10, loc='left')
plt.ylabel('점수', color = 'gray', size=10, loc='bottom')
plt.legend(loc= 'lower left')

# 그래프 이미지로 저장
# plt.savefig('graph2.png',dpi=300)


```
