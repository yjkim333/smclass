```python
import pandas as pd
df = pd.read_excel('score.xlsx',index_col='지원번호')
df
```

```python
df['요일'] = 0
df
```

```python
import random
```

```python
# 컬럼 전체를 수정
df['요일 '] = 5
```

```python
# 0-6 랜덤 숫자를 요일에 입력

def day(value):
	if value == 0 : return '일요일'
	elif value == 1 : return '월요일'
	elif value == 2 : return '화요일'
	elif value == 3 : return '수요일'
	elif value == 4 : return '목요일'
	elif value == 5 : return '금요일'
	elif value == 6 : return '토요일'

# df.iloc[0,9] = 3
for i in range(8):
	df.loc[f'{i+1}번','요일'] = random.randint(0,6)
df

df['요일'] = df['요일'].apply(day)
df
```

```python
# 결과컬럼 생성
df['결과'] = '불합격'
df
```

```python
# 국어가 60점 이상인 경우 합격으로 변경

filt = df['국어'] >= 60
df.loc[filt,'결과'] = '합격'
df
```

```python
# 학교에 고등학교 추가
df['학교'] = df['학교']+'등학교'
df
```

```python
# 키에 cm 추가

# astype(str) 로 문자열로 타입변경 후, 문자열+문자열
df['키'] = df['키'].astype(str)

# astype() -> 컬럼전체 변경할때 에러가 덜 남
# str() -> 특정 컬럼 하나 변경할때 에러가 덜 남
```

```python
df['키'] = df['키'] + 'cm'
df
```

```python
# 학교 그룹

df.groupby('학교').value_counts()
df.groupby('학교').count()


df.groupby('학교')['이름'].count()
```

```python
# mean(),max(),min(),std() -> 컬럼이 숫자가 아니면 실행이 안됨.
df.groupby('학교')['국어'].mean()
df.groupby('학교')['국어'].max()
df.groupby('학교')['국어'].sum()

df.groupby('학교')['국어'].sum().sort_values(ascending=False)
```

```python
# 학년 컬럼 추가
df['학년'] = [3,3,2,1,1,3,2,2]
df
```

```python
# agg()
df.groupby(['학교','학년'])['국어'].agg(['mean','min','max','sum'])
```

```python
df = pd.read_csv('drinks.csv')
df
```

```python
df.shape
```

```python
df.info()

#  4   total_litres_of_pure_alcohol  193 non-null    float64
#  5   continent                     170 non-null    object 
# continent 에 null 값이 23개 있다 파악..
```

```python
# 모든 술의 총량 - total_servings -> beer spirit wine
# => 전체 알코올의 도수 확인

df['total_servings'] = df['beer_servings'] + df['spirit_servings'] + df['wine_servings']
df
```

```python
# 대륙별로 total_servings 를 구하시오
df['continent'].unique()		# NaN까지 5개

# continent NaN 데이터 확인
filt = df['continent'].isnull()
df.loc[filt,'total_servings']
df.loc[filt,'total_servings'].sum()

# continent NaN 데이터 -> 'OT'로 변경
# df['continent'] = df['continent'].fillna('OT')
df['continent'].fillna('OT',inplace=True)
```

```python
df.groupby('continent')['total_servings'].sum()
df.groupby('continent')['total_servings'].sum().sort_values()
```

```python
# 전체 술 소비량 대비 알콜비율이 어느 정도 되는지 출력
# 알콜 비율(total_litres_of_pure_alcohol)을 계산해서 컬럼(alcocho_rate)을 추가 - total_litres_of_pure_alcohol/total_servings
# 전체 술 소비량 대비 알콜비율 높은 나라 순 20개 출력

# 대륙별 알콜비율 계산해서 출력
```

```python
df['alcohol_rate'] = df['total_litres_of_pure_alcohol']/df['total_servings']
df

df['alcohol_rate'].fillna(0,inplace=True)
df

df.sort_values('alcohol_rate',ascending=False).head(20)
df[['country','alcohol_rate']].sort_values('alcohol_rate',ascending=False).head(20)
```

```python
df.groupby('continent')['alcohol_rate'].sum().sort_values()
```

```python
df[['country','alcohol_rate']].sort_values('alcohol_rate',ascending=False).head(10)
```
