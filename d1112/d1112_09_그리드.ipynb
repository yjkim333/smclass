```python

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
# 한글 설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic'		# window용
# matplotlib.rcParams['font.family'] = 'AppleGothic'			# mac용
# fontsize
matplotlib.rcParams['font.size'] = 13
# 그래프에 - 보이기
matplotlib.rcParams['axes.unicode_minus'] = False
```

plt.grid() - 그래프 그리드(안내선) 출력

```python
plt.plot(df['이름'],df['키'],marker='o')

plt.xticks(rotation=45)

# 그리드(안내선) 그리기 - plt.grid()
# axis = x축,ycnr 그리드
# linestyle
# linewidth - 선두께
# color - 선 색상
# alpha - 선 투명도
plt.grid(axis='y', color='pink', ls=':', linewidth=2)
plt.grid(axis='x', alpha=0.2)

# 그래프 값 입력
for i in range(df.shape[0]):
	plt.text(df['이름'][i],df['키'][i]+0.8,df['키'][i],ha='center')


plt.show()
```
