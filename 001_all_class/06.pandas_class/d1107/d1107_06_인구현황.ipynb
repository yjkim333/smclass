```python
import pandas as pd
```

```python
# 상단 3개 삭제 후 가져오기
# index 행정기관으로 지정해서 출력
# 위 5개만 출력


### usecols - 원하는 컬럼만 가져올수 있음 'B,D,E' / 'B,E:Y' / 'A:E'
df = pd.read_excel('연령별인구현황01.xlsx',skiprows=3,index_col='행정기관',usecols='B,E:Y')
df.head()
```

```python
df['0~4세']
```

```python
# df[이름] -> 컬럼을 보는 명령어
# df[슬라이싱]  -> row를 보는 명령어
# df.iloc[번호] -> row를 보는 명령어
# df.loc[이름] -> row를 보는 명령어
df[0:1]
df.iloc[0]
# 데이터 replace 함수를 적용, 타입변경 - astype
# df.iloc[0].str.replace(",","").astype(int)
# df.iloc[0] = df.iloc[0].str.replace(",","").astype(int)
df.describe()
df
```

```python
# 여자부분만 가져오시오
# 위 3개 제외하고, 행정기관 index 지정, 여자부분만 가져오시오.
df_w = pd.read_excel('연령별인구현황01.xlsx',skiprows=3,index_col='행정기관',usecols='B,AB:AV')
df.columns
# 여자컬럼과 남자컬럼 명이 다르기에 통일 시킴
df_w.columns = df.columns
df_w.columns

# 천단위표시,int타입변경
df_w.iloc[0] = df_w.iloc[0].str.replace(",","").astype(int)

```
