```python
# 데이터 추가
df['합계'] = df['국어']+df['영어']+df['수학']+df['과학']+df['사회']
df['결과'] = 'Fail'
df['평균'] = df['합계']/5
df
```

컬럼 순서 변경

```python
df.columns
# pandas-Series - 리스트구조
# 컬럼 순서 변경은 리스트 타입으로 해야됨.
# 그래야 리스트+리스트 할 수 있음
```

```python

# 컬럼을 list 타입으로 변경
cols = list(df.columns)
type(cols)    # -> list
cols

cols[1:-1]  # 여러개 - list
cols[1]  # 1개 - 문자열


# df[df.columns]
df[cols]
# df['이름','평균',...]  <- [] 안에 리스트타입만 올수 있음
# [cols[-1]] cols[1:-1]
df[[cols[0]]+[cols[-1]]+cols[1:-1]]
df = df[[cols[0]]+[cols[-1]]+cols[1:-1]]
df
```
