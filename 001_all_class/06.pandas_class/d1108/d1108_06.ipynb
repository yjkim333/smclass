```python
import pandas as pd
```

```python
df = pd.read_excel('score.xlsx',index_col='지원번호')
df
```

```python
# 학교 이름 변경 - 구로고 -> 가산고 / 디지털고 -> 단지고
# SW특기 변경 - 소문자로 변경
# 4번 학생 SW특기 변경 - django
# 5번 학생 SW특기 변경 - python
# 합계 컬럼 추가
# 결과 컬럼 추가 -> 270 점 이상 Pass 아니면 Fail
```

```python
df['학교'].replace({'구로고':'가산고','디지털고':'단지고'},inplace=True)
df

df['SW특기'] = df['SW특기'].str.lower()
df

df.loc[['4번','5번'],'SW특기'] = 'django','python'
df

df['합계'] = df['국어']+df['영어']+df['수학']+df['과학']+df['사회']
df

df['결과'] = 'Fail'
filt = df['합계'] >= 270
df.loc[filt,'결과'] = 'Pass'
df
```

```python
# 문자열 = 문자열+문자열
df['학교'] = df['학교']+'등학교'
df
```

```python
# #  df['키'] + 'cm' ->
# # str(df.loc['1번','키']) + 'cm'

# # df['키'].astype(str) + 'cm'
df['키'] = df['키'].astype(str) + 'cm'
df
```
