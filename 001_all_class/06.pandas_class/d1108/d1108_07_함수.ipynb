```python
import pandas as pd
```

```python
df = pd.read_excel('score.xlsx',index_col='지원번호')
df
```

```python
def add_cm(height):
	return str(height) + 'cm'
```

```python
# 함수호출
# .apply()
```

```python
df['키'] = df['키'].apply(add_cm)
df
```

```python
df['SW특기'] = df['SW특기'].str.lower()
df
```

```python
def change(lang):
	if pd.notnull(lang):
		return lang.upper()    # lang <- df['SW특기'] 데이터 리스트에서 str로 하나하나 떼어옴.
	else:
		return 'PYTHON'
```

```python
df['SW특기'].apply(change)
```
