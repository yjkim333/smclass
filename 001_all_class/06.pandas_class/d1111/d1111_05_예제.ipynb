```python
import pandas as pd
data = {
   '영화' : ['명량', '극한직업', '신과함께-죄와 벌', '국제시장', '괴물', '도둑들', '7번방의 선물', '암살'],
   '개봉 연도' : [2014, 2015, 2017, 2014, 2015, 2017, 2013, 2015],
   '관객 수' : [1761, 1626, 1441, 1426, 1301, 1298, 1281, 1270], # (단위 : 만 명)
   '평점' : [8.88, 9.20, 8.73, 9.16, 8.62, 7.64, 8.83, 9.10]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
# 관객수 * 10000 -> 관객수로 역순정렬하시오.

df['관객 수'] = df['관객 수']*10000
df
```

```python
df.sort_values('관객 수',ascending=False)
```

```python
# 개봉연도에 연도 를 추가하시오
df['개봉 연도'] = df['개봉 연도'].astype(str)+'년도'
df
```

```python
# 컬럼 삭제
# df.drop(columns='컬럼명',inplace=True)
```

```python
### 천 단위 표시 ###
### 관객수 천단위 표시해서 출력하시오. ###

# df['관객 수'] = df['관객 수'].map('{:,d}'.format)
# df
```

```python
df['관객 수'] = '$'+df['관객 수'].map('{:,d}'.format)
df
```

```python
# $ 없애고 천단위표시 없애고 int 타입으로 변경

df['관객 수'] = (df['관객 수'].str[1:].str.replace(",","").astype(int))/10000
df
```

```python
df.info()
```

```python
# 추천점수 컬럼 추가 - 추천점수 = (관객수*평점)//100
# // -> 몫
df['추천점수'] = (df['관객 수']*df['평점'])//100
df
```

```python
# 연도별 그룹핑 해서 관객수 출력하시오.
df.groupby('개봉 연도')['관객 수'].sum().sort_values(ascending=False)
```
