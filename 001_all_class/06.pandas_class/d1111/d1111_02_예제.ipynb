```python
# 함수 사용 item_price 에 $ 없애고 타입을 float으로 변경
# df['item_price'].str[1:].astype(float)

def func(x):
	return float(x[1:])

df['item_price'] = df['item_price'].apply(func)

# lambda 함수 적용
# df['item_price'] = df['item_price'].apply(lambda x:float(x[1:]))

df
```
