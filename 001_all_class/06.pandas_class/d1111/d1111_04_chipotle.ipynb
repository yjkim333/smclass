```python
# 각 아이템별 가격을 구하시오.
# 주문이 1개인 메뉴를 먼저 출력하시오.
filt = df['quantity']==1 
# df[df['quantity']==1][['item_name','item_price']].min()

df[filt].groupby('item_name')[['item_price']].min().sort_values('item_price').count()
df[filt].groupby('item_name')[['item_price']].min().sort_values('item_price',ascending=False)
```

```python
## 주문당 20달러 이상 주문한 order_id 출력
df_s = df.groupby('order_id')['item_price'].sum().sort_values(ascending=False)
filt = df.groupby('order_id')['item_price'].sum().sort_values(ascending=False)>=20

df_s.loc[filt]


```
