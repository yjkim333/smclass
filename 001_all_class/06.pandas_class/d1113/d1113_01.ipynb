```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
# 한글 설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic'		# window용
# matplotlib.rcParams['font.family'] = 'AppleGothic'			# mac용
# fontsize
matplotlib.rcParams['font.size'] = 13
# 그래프에 - 보이기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_excel('score.xlsx',index_col='지원번호')
df
```

```python
plt.title('학생성적그래프',size = 10)

plt.plot(df.index,df['국어'], label='국어')
plt.plot(df.index,df['영어'], label='영어')
plt.plot(df.index,df['수학'], label='수학')

for i in range(8):
	plt.text(df.index[i],df['국어'][i]+1,df['국어'][i],ha='center')
	plt.text(df.index[i],df['영어'][i]+1,df['영어'][i],ha='center')
	plt.text(df.index[i],df['수학'][i]+1,df['수학'][i],ha='center')

plt.legend()

plt.show()
```

```python
def n_func(x):
	if x-2 > 0:
		return x
	else:
		return 0

n_list = [10,5,-1,6,7]
for i in n_list:
	print('데이터 값 :',n_func(i))

# apply - 함수호출 => pandas에서 적용
```

```python
def func(x):
	if x-30 > 15:
		return x-30
	else : return 0
```

```python
# 누적막대그래프

plt.figure(figsize=(8,4))
plt.title('학생성적그래프',size=12)

plt.bar(df['이름'],df['국어'],width=0.6,label='국어')
plt.bar(df['이름'],df['영어'],bottom=df['국어'],width=0.6,label='영어')
plt.bar(df['이름'],df['수학'],bottom=df['국어']+df['영어'],width=0.6,label='수학')

plt.legend()
# plt.legend(ncol=3)


for i in range(8):
	plt.text(df['이름'][i],func(df['국어'][i]),df['국어'][i],ha='center')
	plt.text(df['이름'][i],df['국어'][i]+df['영어'][i]-20,df['영어'][i],ha='center')
	plt.text(df['이름'][i],df['국어'][i]+df['영어'][i]+df['수학'][i]-10,df['수학'][i],ha='center')

plt.show()
```
