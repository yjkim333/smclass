pie(원)그래프

```python
## 원그래프
values = [30,25,20,13,10,2]
explode = [0.05,0.03,0,0,0,1]  # 시계 반대 방향으로 돔
labels = ['python','java','javascript','oracle','html','etc']
colors = ['y','g','peachpuff','khaki','skyblue','bisque']

plt.title('프로그램 사용률')

# pie(데이터값,labels=라벨값, explode=explode값)
# labels - 글자 입력
# explode - 원그래프 분리
# counterclock = True - 시계반대 방향으로 
# counterclock = False - 시계방향으로 
# startangle = 90 - 12시 방향부터 시작
# autopct = '%.1f' - 그래프에 데이터 출력 - 00.0
# autopct = '%.1f%%' - 그래프에 퍼센트 값으로 데이터 출력 - 00.0%
# colors = 색상설정
plt.pie(values, labels=labels, explode=explode, autopct='%.1f%%',counterclock=False, startangle=90, colors=colors)

plt.legend(loc=(1.1,0), fontsize=10, title='언어별 선호도')

plt.show()
```
