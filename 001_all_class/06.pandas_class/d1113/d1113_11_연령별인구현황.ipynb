```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
# 한글 설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic'		# window용
# matplotlib.rcParams['font.family'] = 'AppleGothic'			# mac용
# fontsize
matplotlib.rcParams['font.size'] = 13
# 그래프에 - 보이기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_excel('연령별인구현황_202410.xlsx',skiprows=3,index_col='행정기관',usecols='b,e:y',nrows=1)
df = df.T

df.columns = ['전국']

df['전국'] = df['전국'].str.replace(',',"").astype(int)

df
```

```python


plt.figure(figsize=(14,5))
plt.bar(df.index,df['전국'])
plt.xticks(rotation=45)

plt.show()
```

```python
df2 = pd.read_excel('연령별인구현황_202410.xlsx',skiprows=3,index_col='행정기관',usecols='b,ab:av',nrows=1)
df2 = df2.T
df2.columns = ['전국']
df2['전국'] = df2['전국'].str.replace(',',"").astype(int)
```

```python
df['전국'].mean()
df.index
```

```python
plt.plot([df['전국'].mean()//1000]*len(df.index),df.index)
```

```python
plt.figure(figsize=(14,7))
plt.barh(df.index,-df['전국']//1000,label='남')
plt.barh(df.index,df2['전국']//1000,label='여')

plt.xlabel('성별')
plt.ylabel('연령대')

plt.plot([-df['전국'].mean()//1000]*len(df.index),df.index,label='남 평균',ls=':',alpha=0.8)
plt.plot([df2['전국'].mean()//1000]*len(df.index),df.index,label='여 평균',ls=':',alpha=0.8)

plt.legend(fontsize=11)

plt.show()
```
