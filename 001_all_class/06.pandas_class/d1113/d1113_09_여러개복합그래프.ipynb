```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
# 한글 설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic'		# window용
# matplotlib.rcParams['font.family'] = 'AppleGothic'			# mac용
# fontsize
matplotlib.rcParams['font.size'] = 13
# 그래프에 - 보이기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_excel('score.xlsx',index_col='지원번호')
df
```

복합그래프 - 여러개 그래프

```python
# 복합그래프 - 여러개 그래프
# plt.subplots(2,2) -> 2x2 그래프
# 

# 4개(2*2) 그래프 혼합해서 사용
fig,axs = plt.subplots(2,2,figsize=(15,14))

# 메인 타이틀 - suptitle()
fig.suptitle('4개 복합 그래프')

# 4개 그래프 => 1번째 그래프 - axis[0,0] / 2 - axis[0,1] / 3 - axis[1,0] / 4 - axis[1,1] 
# 설정 - set_명령어

## 1번째 그래프 [0,0] - bar graph
axs[0,0].bar(df['이름'],df['국어'],label='국어',color='tan')

axs[0,0].set_title('1. 국어점수 그래프',size=13)
# x축, y축 이름 - .set(xlabel='이름',ylabel='점수')
axs[0,0].set(xlabel='이름',ylabel='점수')
# 배경색 - .set_facecolor('ivory')
axs[0,0].set_facecolor('ivory')


# 범례
axs[0,0].legend(loc='upper center')


## 2번째 그래프 [0,1] - plot graph
axs[0,1].plot(df['이름'],df['영어'],label='영어')
axs[0,1].plot(df['이름'],df['수학'],label='수학')

axs[0,1].set_title('2. 영어,수학 점수 그래프',size=13)
axs[0,1].set(xlabel='이름',ylabel='점수')

axs[0,1].legend()


## 3번째 그래프 [1,0] - barh graph
axs[1,0].set_title('''3. '키' 가로바 그래프''',size=13)
axs[1,0].barh(df['이름'],df['키'],height=0.6)
axs[1,0].set_xlim(140,210)


for i in range(8):
	axs[1,0].text(df['키'][i]+1,df['이름'][i],df['키'][i])


## 3번째 그래프 [1,1] - plot graph
axs[1,1].set_title('4. 사회점수 선 그래프',size=13)
axs[1,1].plot(df['이름'],df['사회'],color='khaki')

def s_func(x):
	if x <=35:
		return x-3
	return x+2

for i in range(8):
	axs[1,1].text(df['이름'][i],s_func(df['사회'][i]),df['사회'][i],ha='center')




plt.show()
```
