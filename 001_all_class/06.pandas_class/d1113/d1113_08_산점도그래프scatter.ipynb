```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
# 한글 설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic'		# window용
# matplotlib.rcParams['font.family'] = 'AppleGothic'			# mac용
# fontsize
matplotlib.rcParams['font.size'] = 13
# 그래프에 - 보이기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_excel('score.xlsx',index_col='지원번호')
df
```

```python
df['학년'] = [3,3,2,1,1,3,2,2]
df
```

산점도 그래프

```python
# 산점도 그래프
# plt.scatter()

# sizes = np.random.rand(8)*1000
sizes = df['학년']*200

plt.title('scatter graph - 산점도 그래프')

plt.xlabel('영어')
plt.ylabel('수학')

# s - 산점도 크기 변경	// c=df['학년'] -> '학년'별로 사이즈 표시
# c - 산점도 색상 표시 // c=df['학년'] -> '학년'별로 색상 표시
# cmap - 색상표 사용
plt.scatter(df['영어'],df['수학'],marker='o',s=sizes, c=df['학년'], cmap='Paired' , alpha=0.5)

# colorbar - 색상바 생성 - 산점도 색상 설명
# ticks - 색상바 범위
# label - 색상바 설명
# shrink - 색상바 크기
# orientation - 색상바 위치 - horizontal / vertical
plt.colorbar(ticks=[1,2,3],label='학년',shrink=0.5, orientation='horizontal')

plt.show()
```
