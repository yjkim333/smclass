```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
# 한글 설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic'		# window용
# matplotlib.rcParams['font.family'] = 'AppleGothic'			# mac용
# fontsize
matplotlib.rcParams['font.size'] = 13
# 그래프에 - 보이기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_excel('출산율2.xlsx',skiprows=2,skipfooter=3,index_col=0)
df = df.T
df.columns = ['출생아수']
df
```

```python
plt.figure(figsize=(14,5))
plt.title('연도별 출생아수',size=12)
plt.plot(df.index,df['출생아수'],color='m',marker='o',mec='m',mfc='w',ms=8)
plt.show()
```
