```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
# 한글 설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic'		# window용
# matplotlib.rcParams['font.family'] = 'AppleGothic'			# mac용
# fontsize
matplotlib.rcParams['font.size'] = 13
# 그래프에 - 보이기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
data = {
   '영화' : ['명량', '극한직업', '신과함께-죄와 벌', '국제시장', '괴물', '도둑들', '7번방의 선물', '암살'],
   '개봉 연도' : [2014, 2019, 2017, 2014, 2006, 2012, 2013, 2015],
   '관객 수' : [1761, 1626, 1441, 1426, 1301, 1298, 1281, 1270], # (단위 : 만 명)
   '평점' : [8.88, 9.20, 8.73, 9.16, 8.62, 7.64, 8.83, 9.10]
}
df = pd.DataFrame(data)
df
```

```python
# 영화 관객수를 막대 그래프로 출력하시오.
# 평균 막대그래프 추가, 평균 선을 표시

# 영화-평균, 개봉연도-2024 관객수-평균관객수, 평점-평균평점

df['평점'].mean()
df['관객 수'].mean()

df.loc[8] = ['평균',2024,df['관객 수'].mean(),df['평점'].mean()]
df

m_data = [df['관객 수'].mean()]*9

plt.figure(figsize=(14,4))
plt.title('영화 관객수')
plt.grid(axis='y',alpha=0.4,ls=':')

m_bar = plt.bar(df['영화'],df['관객 수'],width=0.5)

plt.xticks(rotation=45)
plt.ylim(1200,1800)

plt.text(df['영화'][0],df['관객 수'][0],df['관객 수'][0])

m_bar[len(df)-1].set_color('tan')

plt.plot(df['영화'],m_data,color='r', ls='--', alpha=0.3)

plt.show()
```
