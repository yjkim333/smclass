```python
import matplotlib.pyplot as plt
import matplotlib
# 한글 설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic'		# window용
# matplotlib.rcParams['font.family'] = 'AppleGothic'			# mac용
# fontsize
matplotlib.rcParams['font.size'] = 13
# 그래프에 - 보이기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
import pandas as pd
import numpy as np
```

```python
# x축,y축 데이터 - list,tuple 타입, DataFrame데이터
x = [1,2,3]
y = [2,4,8]

# 그래프 생성
# plot - 선 그래프
plt.plot(x,y)

# 상단 타이틀, size = 글자크기 변경
plt.title('라인그래프 LINE',size = 10)

# 그래프 하단 X축설명 , color = 글자색 변경, loc = (right,center,left) 글자 위치 변경
plt.xlabel('X축 날짜',color = '#0a8dad',loc='right',size = 12)

# 그래프 좌측 Y축설명, color = 글자색 변경, loc = (top,center,bottom) 글자 위치 변경
plt.ylabel('Y축 매출액',color = '#c72261', loc = 'top')


# 그래프 눈금 단위
plt.xticks([1,2,3])
plt.yticks([2,3,4,5,8,10])


# [<matplotlib.lines.Line2D at 0x1ea46edbec0>] 설명 글자 없애기
plt.show()
```
