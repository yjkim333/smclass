```python
import matplotlib.pyplot as plt
import matplotlib
# 한글 설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic'		# window용
# matplotlib.rcParams['font.family'] = 'AppleGothic'			# mac용
# fontsize
matplotlib.rcParams['font.size'] = 13
# 그래프에 - 보이기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
# 바 크기- width
# 투명도 조절 - alpha
# 라벨변경, 라벨 기울기 설정
```

```python
label = ['홍길동','유관순','강감찬']
values = [180,187,184]

# 그래프 크기 조절
plt.figure(figsize=(3,3))

# 선 그래프
# plt.plot(label,values)

# plt.bar()막대 그래프
# color : 그래프 색상
# width : 바 크기 조절
# plt.bar(label,values, color='green')
# 바 색상 여러개로 처리
colors=['r','g','b']
plt.bar(label,values, color=colors, width=0.2, label='학생키')

# 상단제목
plt.title('Bar Graph',size=10)

# x축 라벨 변경
# rotation - 각도변경
plt.xticks(label,['1번','2번','3번'], rotation=180)
# y축 라벨 변경
plt.yticks([0,50,100,150,200])

plt.legend(loc=(1.05,0))

plt.show()
```

set_hatch() - 막대그래프 무늬 설정

```python
colors = ['red','green','blue']
bar = plt.bar(label,values,width=0.3,color=colors,alpha=0.5)

# 막대그래프의 무늬 설정
bar[0].set_hatch('/')
bar[1].set_hatch('x')
bar[2].set_hatch('..')
```

* 범위 설정
* xticks / yticks  => 0 부터 시작
* xlim() / ylim() => 특정 범위 지정

```python
colors = ['red','green','blue']
bar = plt.bar(label,values,width=0.3,color=colors,alpha=0.5)

# 범위지정
# plt.yticks([0,100,150,200])			# 0부터 시작
plt.ylim(175,190)		# 특정 범위 지정

plt.show()
```

barh - 옆으로 누운 막대그래프

```python
label = ['홍길동','유관순','강감찬']
values = [180,187,184]

colors = ['red','green','blue']
plt.barh(label,values,color=colors,height=0.3)

plt.xlim(175,190)

```
