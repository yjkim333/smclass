```python
import matplotlib.pyplot as plt
import matplotlib
# 한글 설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic'		# window용
# matplotlib.rcParams['font.family'] = 'AppleGothic'			# mac용
# fontsize
matplotlib.rcParams['font.size'] = 13
# 그래프에 - 보이기
matplotlib.rcParams['axes.unicode_minus'] = False
```

그래프 데이터 값 표시 - plt.text()

```python
# x축,y축 데이터 - list,tuple 타입, DataFrame데이터
x = [1,2,3]
y = [2,4,8]

plt.figure(figsize=(8,5))
plt.plot(x,y, label='범례데이터', marker='o')

# 그래프에 데이터값 넣기
# text(x좌표위치,y좌표위치,데이터값)
# ha - 글자 정렬 위치
# plt.text(x[0],y[0]+0.2,2,ha='center')
# plt.text(x[1],y[1]+0.2,y[1],ha='center')

## plt.text()
# for문 - 그래프에 데이터값 넣기
for idx,txt in enumerate(y):
	plt.text(x[idx]+0.05,y[idx]-0.3,y[idx],ha='center',color = 'green')

plt.title('Line graph')
plt.legend()
plt.show()
```
