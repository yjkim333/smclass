```python
# x축,y축 데이터 - list,tuple 타입, DataFrame데이터
x = [1,2,3]
y = [2,4,8]

# 그래프 생성
# plot - 선 그래프
plt.plot(x,y)

# 상단 타이틀
plt.title('라인그래프 LINE')

# [<matplotlib.lines.Line2D at 0x1ea46edbec0>] 설명 글자 없애기
plt.show()
```
