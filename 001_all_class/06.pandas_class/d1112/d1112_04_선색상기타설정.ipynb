```python
import matplotlib.pyplot as plt
import matplotlib
# 한글 설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic'		# window용
# matplotlib.rcParams['font.family'] = 'AppleGothic'			# mac용
# fontsize
matplotlib.rcParams['font.size'] = 13
# 그래프에 - 보이기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
# x축,y축 데이터 - list,tuple 타입, DataFrame데이터
x = [1,2,3]
y = [2,4,8]
yy = [1,8,3]

# 그래프 생성
# plot - 선 그래프
# linewidth - 그래프 선 두께
# color - 그래프 선 색상
# linestyle - 축약(ls) 라인스타일 설정 = ':'-점선 / '-' - 실선 / '--' - 대쉬선 / '-.' - -.선
# alpha - 그래프 투명도 설정

# 마커
# marker = 'x'/'o'/'v' / markersize = 10 / markeredgecolor - 마커테두리 / markerfacecolor - 마커색상

# plt.plot(x,y, label='인원변동', linewidth = 5, marker = 'o', markersize = 15, markeredgecolor = 'black',markerfacecolor = 'white')
plt.plot(x,y, label='인원변동', linewidth = 5, marker='o', ms=15, mec = 'black',mfc = 'white')
# plt.plot(x,yy, label='비용변동', color:'pink', marker = 'v', markersize = 15, alpha = 0.4, ls = '-.', mfc = 'red')

# 선 설정 단축 -> 'bv-:' => b - 선색상 / v - 마커표시,/ : - 선모양
plt.plot(x,yy, 'bv-.', label='비용변동')

# 상단 타이틀, size = 글자크기 변경
plt.title('라인그래프 LINE',size = 10)

# 범례 - 그래프 선 설명문구, 범례위치 - loc='upper right','upper center','lower center','center right','lower right','center'...
# plt.legend(loc='lower right')
plt.legend(loc= (1.02,0))
```
