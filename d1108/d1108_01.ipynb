```python
# startswith() - 해당 값으로 시작되는 단어 검색
filt = df['이름'].str.startswith('김')
df[filt]

# contains() - 해당 값이 포함되어 있는 단어 검색
filt = df['이름'].str.contains('수')
df[filt]
df.loc[filt,['이름','학교','키']]


# isin(리스트) - 리스트에 있는 값이 포함이 되어 있는지 확인 / 리스트타입만 가능
langs = ['Python','Java']
filt = df['SW특기'].isin(langs)
df[filt]

# lower() - 해당 값을 모두 소문자로 치환
langs = ['python','java']
filt = df['SW특기'].str.lower().isin(langs)
df[filt]

# 포함이 되어 있으면 모두 검색
# - NaN 데이터 처리해야함
filt = df['SW특기'].str.lower().str.contains('j',na=False)
# filt
df[filt]
```

```python
import numpy as np
```

```python
# fillna() - 모든 컬럼의 NaN 데이터를 모두 변경
# fillna() - inplaca=True 해야 적용됨.
df.fillna('데이터 없음',inplace=True)
df
```

```python
# dropna()
# axis - 축방향으로 삭제
# how = 'any' - NaN이 1개라도 있으면 해당되는 row 전체나 해당되는 column 전체 삭제
# how = 'all' - row나 column에서 해당되는 데이터가 모두 NaN이면 그 row나 column 삭제
# axis = 'index' - row방향삭제 / axis = 'columns' -columns방향삭제 , how = 'any' / how = 'all'
# 디폴트로 axis='index' how = 'any'
# inplace = True
df.dropna()
```

```python
# np.nan - NaN 데이터 값으로 수정입력
df['학교'] = np.nan
df
```

```python
df.dropna(axis='index',how='all')		 # row 방향으로 모든 값이 NaN인 row 삭제
df.dropna(axis='columns',how='all')	 # column 방향으로 모든 값이 NaN인 row 삭제 -> '학교' 컬럼 삭제
df.dropna(axis='index',how='any')		 # NaN 값이 1개라도 있는 row 는 모두 삭제
df.dropna(axis='columns',how='any')		 # NaN 값이 1개라도 있는 column 은 모두 삭제 -> '학교''SW특기' 삭제
# df
```
