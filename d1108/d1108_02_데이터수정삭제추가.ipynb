```python
import pandas as pd
```

```python
df = pd.read_excel('score.xlsx',index_col='지원번호')
df
```

데이터 수정

```python
# 데이터 수정
# repalce({:}) - 해당되는 값 -> 다른 값
# inplace=True

df['학교'].replace({"구로고":"가산고"},inplace=True)
# 여러개 동시 변경
df['학교'].replace({'구로고':'가산고','디지털고':'단지고'})
df
```

```python
# 특정컬럼 값 변경
# 컬럼 = 컬럼.변경값
df['SW특기'].str.lower()
df['SW특기'].str.upper()
df['SW특기'] = df['SW특기'].str.upper()
df['SW특기'] = df['SW특기'].str.lower()
df

# 첫글자만 대문자로 변경
df['SW특기'].str.capitalize()

# 값 입력해서 변경
df['키'] = 200
df

# ----------------------------

# 특정 셀 값 변경
df.loc['1번','국어'] = 100
df.loc['2번',['국어','영어']] = 10, 20
df.loc[['2번','3번'],['국어','영어']] = 10, 20
df

# 컬럼명 = 문자+문자
df['학교'] = df['학교']+'등학교'
df
```

데이터 추가
- 컬럼 추가 - df[추가할 컬럼명]
- row 추가 - df.loc[추가할 index명]

```python
# 데이터 추가 방법

# 컬럼에 없는 컬럼 이름을 입력하면 컬럼이 추가됨.
df['합계'] = 0
df['합계'] = df['국어']+df['영어']+df['수학']+df['과학']+df['사회']
df['평균'] = df['합계']/5
df
df['결과'] = 'Fail'
df
```

```python
df.loc['1번','결과']
```

조건식을 사용해서 값 변경

```python
# 결과컬럼 추가
df['결과'] = 'Fail'

# 결과컬럼에 조건식을 사용하여 값을 변경
filt = df['합계'] >= 400
df.loc[filt,'결과'] = 'Pass'
df

# '합산'에 국어점수가 90점 이상이면 +100 점
df['합산'] = 100
df
filt = df['국어'] >= 90
df.loc[filt,'합산'] = df['합산']+100
df
```

데이터 삭제
- columns
- index

```python
# 컬럼 삭제
# drop(columns=)
# inplace=True

df.drop(columns='합산', inplace=True)
df

# 컬럼 2개 삭제
df.drop(columns=['국어','영어'])

# 컬럼 전체(<-리스트구조)에 슬라이싱으로 특정 컬럼을 삭제
df.columns[3:5]
df.drop(columns=df.columns[3:5])
```

```python
# index 삭제 (row 값 삭제)
# drop(index=)
# inplace=True

df.drop(index='1번')
df
```

```python
filt = df['수학'] <= 80
filt

# 조건식이 적용된 컬럼의 index 값
df[filt].index
```

```python
# 조건식을 이용한 삭제

# 수학 점수가 80점 이하인 학생 데이터 삭제
filt = df['수학'] <= 80
filt
df.drop(index = df[filt].index)

# '김'이 포함되어 있는 학생을 삭제하시오.
filt = df['이름'].str.contains('김')
df.drop(index = df[filt].index)
```

row 추가

```python
# row 추가
# index 이름 추가해서 입력하면 추가됨.
# 데이터는 list 로 입력.
df.loc['9번'] = ['유승찬','디지털고등학교',184,100,100,100,100,100,'Django',100*5,(100*5)/5,'Pass']
df
```
