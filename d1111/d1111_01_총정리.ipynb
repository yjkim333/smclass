```python
df = pd.read_excel('score.xlsx',index_col='지원번호')
df
```

```python
# 국어 50 점 이하 학생에 점수 +20

# 조건식
# filt = df['국어'] <= 50
# df.loc[filt,'국어'] = df['국어'] +20
# df

# ---------------------------------------------------

# 함수 - if문을 쓸 수 있음
# 함수선언
def func(value):
	if value < 50:
		return value +20
	else:
		return value

# 함수호출 - .apply() 사용
df['국어'] = df['국어'].apply(func)
df
```

```python
# 국어 50 점 이하 학생에 점수 +20 / 50점 이상 +10

# 조건식
# filt = df['국어'] <= 50
# df.loc[filt,'국어'] = df['국어'] +20
# filt = df['국어'] > 50
# df.loc[filt,'국어'] = df['국어'] +10
# df


# 함수 - if문을 쓸 수 있음
# 함수선언
def func(value):
	if value < 50:
		return value +20
	elif value > 10:
		return value +10
	else:
		return value

# 함수호출 - .apply() 사용
df['국어'] = df['국어'].apply(func)
df
```
