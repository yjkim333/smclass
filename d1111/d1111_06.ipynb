```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('chipotle.tsv',sep='\t')
df
```

```python
# 메뉴별 주문 건수를 출력하시오
df.groupby('item_name')['quantity'].count()
```

```python
# 메뉴별 총 주문수량을 출력하시오
df.groupby('item_name')['quantity'].sum()
```

```python
# 주문당 평균 금액 출력
# df.info()
df['item_price'] = df['item_price'].str[1:].astype(float)
```

```python

df.groupby('order_id')['item_price'].sum().mean()
```

```python
# 아이템별 가격 금액을 출력하시오
df.groupby('item_name')['item_price'].min().sort_values(ascending=False)
```

```python
# 주문별, 높은 가격순으로 주문개수를 출력하시오.
df.groupby('order_id')[['quantity','item_price']].sum().sort_values('item_price',ascending=False)
```
