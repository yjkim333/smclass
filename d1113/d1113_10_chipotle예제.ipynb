```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
# 한글 설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic'		# window용
# matplotlib.rcParams['font.family'] = 'AppleGothic'			# mac용
# fontsize
matplotlib.rcParams['font.size'] = 13
# 그래프에 - 보이기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_csv('chipotle.tsv',sep='\t')
df
```

```python
# 메뉴별로 주문개수 5개만 뽑아서(head()) 막대 그래프로 출력
```

```python
df.groupby('item_name')['quantity'].sum()
df.groupby('item_name')['quantity'].sum().index

f_df = df.groupby('item_name')['quantity'].sum().head(10)

# y = df.groupby('item_name')['quantity'].sum().head(10)
# x = df.groupby('item_name')['quantity'].sum().head(10).index
df.groupby('item_name')['quantity'].sum().mean()

f_df.loc['평균'] = df.groupby('item_name')['quantity'].sum().mean()

x = f_df.index
y = f_df
```

```python
plt.figure(figsize=(10,5))

f = plt.bar(x,y,width=0.6)
plt.xticks(rotation=45)

for i in range(11):
	plt.text(x[i],y[i]+5,int(y[i]),ha='center')

f[len(x)-1].set_color('skyblue')

plt.plot(x,[df.groupby('item_name')['quantity'].sum().mean()]*11,color='red',ls=':')

plt.ylim(0,240)

plt.grid(ls=':',alpha=0.5)
plt.show()
```
