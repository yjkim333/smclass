```python
## 원그래프
values = [30,25,20,13,10,2]
explode = [0.1,0,0,0,0,0]  # 시계 반대 방향으로 돔

wedgeprops = {'width':0.4,'edgecolor':'white','linewidth':2}
# wedgeprops=
# 'width':0.4 - 중간 공백 크기 - 도넛
# 'edgecolor':'white' - 테두리색
# 'linewidth':2 - 테두리 두께

labels = ['python','java','javascript','oracle','html','etc']
colors = ['y','g','peachpuff','khaki','skyblue','bisque']

# 데이터 값이 2% 이하는 출력되지 않도록 함수 적용
# pct - values의 퍼센트 값 생성
# def custom_autopct(pct):
# 	if pct >= 10:
# 		# return ('%.1f%%' % pct)
# 		# return ('{:.1f}%'.format(pct))
# 		return (f'{pct:.1f}%')
# 	return ''

def custom_autopct(pct):
	return f'{pct:.1f}%' if pct >=10 else ''


# pctdistance= - 글자 위치
# wedgeprops= - 중간 공백 - 도넛
plt.pie(values, labels=labels, colors=colors, startangle=90, counterclock=False, autopct=custom_autopct,wedgeprops=wedgeprops,\
				pctdistance=0.8)

plt.show()
```
