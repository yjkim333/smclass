```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
# 한글 설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic'		# window용
# matplotlib.rcParams['font.family'] = 'AppleGothic'			# mac용
# fontsize
matplotlib.rcParams['font.size'] = 13
# 그래프에 - 보이기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_excel('score.xlsx',index_col='지원번호')
df
```

```python
df['키'] = df['키'].astype(str)+'cm'
```

```python
def cm_func(x):
	return str(x)+'cm'

# df['키'] -> 배열타입
# => 배열타입의 모든 것에 함수적용 => apply(함수명)
df['키'] = df['키'].apply(cm_func)
df
```

```python
def k_func(x):
	if x < 50:
		return 50
	return x


# int,str,float - 단일 데이터 값 => 함수명(데이터값)
for i in range(8):
	print(k_func(df['국어'][i]))

	
```

```python

plt.bar(df['이름'],df['국어'],label='국어')
plt.bar(df['이름'],df['영어'],bottom=df['국어'],label='영어')
plt.bar(df['이름'],df['수학'],bottom=df['국어']+df['영어'],label='수학')

def func(x):
	return x/2

for i in range(8):
	plt.text(df['이름'][i],func(df['국어'][i]),df['국어'][i],ha='center')
	plt.text(df['이름'][i],func(df['국어'][i]+df['국어'][i]+df['영어'][i]),df['영어'][i],ha='center')
	plt.text(df['이름'][i],func(df['국어'][i]+df['국어'][i]+df['영어'][i]+df['영어'][i]+df['수학'][i]),df['수학'][i],ha='center')
```
