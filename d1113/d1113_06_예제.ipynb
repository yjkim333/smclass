```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
# 한글 설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic'		# window용
# matplotlib.rcParams['font.family'] = 'AppleGothic'			# mac용
# fontsize
matplotlib.rcParams['font.size'] = 13
# 그래프에 - 보이기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_excel('score.xlsx',index_col='지원번호')
df
```

```python
# 학교별 인원을 원그래프로 출력

df.groupby('학교').size()
df.groupby('학교')['이름'].count()
df['학교'].value_counts()

# explode = [0.02,0]
wedgeprops = {'width':0.6,'edgecolor':'white', 'linewidth':4}

s_df = df.groupby('학교')['이름'].count()
s_df.index  # ['구로고', '디지털고']

plt.pie(s_df,labels=s_df.index, autopct='%.1f%%',startangle=90, wedgeprops=wedgeprops, pctdistance=0.7)

plt.show()
```
