```python
import pandas as pd
import numpy as np
df = pd.read_excel('score.xlsx',index_col='지원번호')
df
```

```python
### 결측치(NaN데이터)

# NaN 데이터 채우기 - fillna()
# df.fillna('',inplace=True)
df.fillna('없음',inplace=True)
df

```

fillna() - NaN 데이터 값 처리

```python
### numpy
# 학교 컬럼에 NaN 데이터를 입력 - np.nan
df['학교'] = np.nan
# df.fillna('없음')  # inplace=True, 모든 NaN 값에 해당 데이터를 입력
df['SW특기'].fillna('없음',inplace=True)  # - 해당 컬럼에만 적용
df
```

```python
df = pd.read_excel('score.xlsx',index_col='지원번호')
```

```python
# axis - 축
# axix = 'index' - index 방향으로 삭제
# -> NaN 데이터가 있는 4,5번 row 데이터가 삭제
df.dropna(axis='index',how='any')
```

```python
df = pd.read_excel('score.xlsx',index_col='지원번호')
df
```

```python
# dropna() -> NaN 데이터 삭제
# axix : 축 - index-row삭제 / columns-column 삭제
# how - default는 any - NaN 값이 1개라도 있으면 삭제 / all - 전부 NaN 일 때 삭제 
df.dropna(axis='columns',how='all')
```

```python
df = pd.read_excel('score.xlsx',index_col='지원번호')
df['학교'] = np.nan
df
```

```python
df.dropna(axis='columns',how='all')
```

* fillna() - NaN 데이터를 ()해당 값으로 채워줌
* dropna() - NaN 데이터를 삭제
- axis : 축 - index -> row 방향 / columns -> column 방향으로 삭제
- how - 어떤 경우에 삭제할지 결정 - any -> row 나 column에 한개라도 NaN 값이 있으면 삭제(default) / all -> 모두 NaN 일때 삭제

* np.nan - numpy 라이브러리 -> NaN 값 입력
